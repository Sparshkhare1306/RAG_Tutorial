```python
sample_texts= {
    "../data/text_files/python_intro.txt": ''' 

Python is a high-level, interpreted programming language known for its readability, simplicity, and broad applicability. Created by Guido van Rossum and first released in 1991, Python has grown into one of the most widely used languages in the world, powering everything from small scripts to large-scale systems.

At its core, Python emphasizes clean and readable code. Its syntax is intentionally straightforward, often resembling plain English, which makes it an excellent choice for beginners while still being powerful enough for professionals. One of its defining features is the use of indentation to define code blocks instead of braces or keywords, which enforces a consistent coding style and improves clarity.

Python is an interpreted language, meaning code is executed line by line rather than compiled into machine code beforehand. This makes development faster and more interactive, though sometimes slightly slower in execution compared to compiled languages like C++. However, this trade-off is often acceptable given Python’s productivity benefits.

The language supports multiple programming paradigms, including procedural, object-oriented, and functional programming. This flexibility allows developers to choose the style that best fits their problem. Python’s dynamic typing system means you don’t need to declare variable types explicitly, which speeds up coding but requires careful handling to avoid runtime errors.

A major strength of Python is its vast standard library and ecosystem of third-party packages. Libraries like NumPy and pandas are widely used in data science, while frameworks such as Django and Flask support web development. In artificial intelligence and machine learning, tools like TensorFlow and PyTorch have made Python a dominant language in the field.

Python is also highly portable. Code written on one platform (Windows, macOS, Linux) can usually run on another with little or no modification. Its large and active community contributes to extensive documentation, tutorials, and open-source projects, making it easier to learn and troubleshoot.

In terms of real-world use cases, Python is used in web development, automation and scripting, scientific computing, data analysis, artificial intelligence, game development, and even cybersecurity. Companies like Google, Netflix, and Instagram rely on Python in various parts of their technology stacks.

Despite its many strengths, Python does have limitations. Its performance is generally slower than lower-level languages, and its dynamic nature can lead to runtime errors that would be caught earlier in statically typed languages. It’s also not typically used for mobile app development or systems programming where performance and hardware control are critical.

Overall, Python stands out as a versatile, accessible, and powerful language that balances ease of use with deep capability. Whether someone is just starting out in programming or building advanced systems, Python offers tools and a community that make it a compelling choice.



''',


"../data/text_files/machine_learning.txt": ''' 
Machine learning, often abbreviated as ML, is a subfield of Artificial Intelligence that focuses on building systems capable of learning from data and improving their performance over time without being explicitly programmed for every task. Instead of following rigid rules, machine learning models identify patterns, relationships, and structures within data to make predictions or decisions.

At the heart of machine learning is the idea of training a model. A model is essentially a mathematical representation that maps inputs (data) to outputs (predictions). During training, the model is exposed to data and adjusts its internal parameters to minimize errors. This process is guided by a loss function, which measures how far the model’s predictions are from the actual outcomes.

There are three main types of machine learning. The first is supervised learning, where the model is trained on labeled data—meaning each input comes with a correct output. Common tasks here include classification (like identifying whether an email is spam) and regression (predicting numerical values such as house prices). Algorithms like Linear Regression and Decision Trees are widely used in this category.

The second type is unsupervised learning, where the data has no labels. The model tries to uncover hidden patterns or groupings within the data. A common example is clustering, where similar data points are grouped together. Techniques like K-means clustering are often used for this purpose.

The third type is reinforcement learning, where an agent learns by interacting with an environment and receiving rewards or penalties based on its actions. Over time, the agent aims to maximize cumulative rewards. This approach is used in areas like robotics, game playing, and autonomous systems.

A key concept in ML is generalization—the ability of a model to perform well on new, unseen data. A model that performs well on training data but poorly on new data is said to suffer from overfitting. The opposite problem, where a model is too simple to capture patterns in the data, is called underfitting. Techniques such as regularization, cross-validation, and proper dataset splitting (training, validation, testing) help manage these issues.

Data plays a central role in machine learning. High-quality, relevant, and well-prepared data often matters more than the choice of algorithm. Preprocessing steps such as cleaning data, handling missing values, and scaling features are essential before training models.

In practice, machine learning relies heavily on tools and frameworks. Libraries like scikit-learn are commonly used for traditional ML tasks, while TensorFlow and PyTorch are popular for deep learning, a specialized area of ML that uses neural networks with many layers.

Machine learning is used across many industries. It powers recommendation systems (like suggesting movies or products), natural language processing (chatbots and translation), computer vision (image recognition), fraud detection in finance, and predictive analytics in healthcare and business.

Despite its power, ML comes with challenges. Models can inherit biases from data, require large amounts of computational resources, and sometimes act as “black boxes,” making their decisions difficult to interpret. Ethical considerations, fairness, and transparency are increasingly important topics in the field.

In summary, machine learning is about teaching computers to learn from data and make intelligent decisions. It combines mathematics, statistics, and computer science to create systems that adapt and improve, making it one of the most influential technologies in the modern world.




'''

}
```

```python
### TextLoader

from langchain_community.document_loaders import TextLoader


loader= TextLoader("../data/text_files/python_intro.txt", encoding='utf-8')

document= loader.load()

print(document)
```

```python
### Directory loader

from langchain_community.document_loaders import DirectoryLoader

dir_loader= DirectoryLoader(
    "../data/text_files",
    glob= "**/*.txt",
    loader_cls= TextLoader,
    loader_kwargs={'encoding':'utf-8'},
    show_progress= False,

)

documents= dir_loader.load()
documents
```

```python
from langchain_community.document_loaders import PyPDFLoader, PyMuPDFLoader

pdf_loader= DirectoryLoader(
    "../data/pdf_files",
    glob= "**/*.pdf",
    loader_cls= PyMuPDFLoader,
    show_progress= False,

)

pdf_documents= pdf_loader.load()
pdf_documents
```
